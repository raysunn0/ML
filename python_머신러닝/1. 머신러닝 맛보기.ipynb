### sklearn.model_selection.train_test_split
* sklearn.model_selection.train_test_split(*arrays, test_size=None, train_size=None, random_state=None, shuffle=True, stratify=None)[source]

> https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html?highlight=train_test_split#sklearn.model_selection.train_test_split

random_state : seed 배정 \
해당 주소로 임시저장, 분리... 그 번호 끄지 않으면 계속 거기로 실행 \
-> train data는 random_state 이용 \
seed를 고정하면 난수 뽑을 때도 일정한 수를 뽑아냄

random_stat=11 : 호출할 때마다 같은 학습/테스트용 데이터 세트를 생성하기 위해 주어지는 난수 발생 값 \
train_test_split()은 호출할 때 무작위로 데이터를 분리하기 때문에, random_state를 지정함으로 수행할 때마다 다른 학습/테스트용 데이터를 만들어줘야함 \
동일한 데이터 세트로 분리해주기 위해 random_state를 일정한 숫자 값으로 부여하면 된다.

```python
from sklearn.model_selection import train_test_split
```

```python
# 80%는 학습데이터로 사용, 20%는 테스트데이터로 사용
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)
# random_state는 고정시키는 것

# X_train은 학습용 피처 데이터 세트
# X_test는 테스트용 피처 데이터 세트
# y_train은 학습용 레이블 데이터 세트
# y_test는 테스트용 레이블 데이터 세트
```

- Train: 모델을 학습시키기 위한 dataset. "모델을 학습하는데에는 Train Dataset만 이용한다"
- Test: 학습과 검증이 검증이 완료된 모델의 성능을 평가하기 위한 dataset
- Validation: 이미 학습된 완료된 모델을 검증하기 위한 dataset으로, Train에서 일부를 가져와 8:2의 비율로 Validation에서 2를 가져오게 됨
> Validation set은 모델을 update, 즉 학습을 시키진 않지만 학습에 '관여'는 하지만, Test set은 학습에 전혀 관여하지 않고 오직 '최종 성능'을 평가하기 위해 쓰이게 된다. (과적합의 문제를 방지하기 위해)

### sklearn.tree.DecisionTreeClassifier
* class sklearn.tree.DecisionTreeClassifier(*, criterion='gini', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)[source]
> https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html?highlight=decisiontreeclass#sklearn.tree.DecisionTreeClassifier

tree에 있다 -> from sklearn.tree import DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)

# 학습 수행
dt_clf.fit(X_train, y_train)
# 왜 2개로 나눌까? input값과 label값 2개가 필요하기 때문
# 지도학습이기 때문에 label 값이 필요한 것
# 뒤의 y_train을 y_test로 하면 X_train과 갯수가 차이나서 오류가 나게 됨

# 학습을 시키는 알고리즘 DecisionTreeClassifier 사용
# 선언하고 전달해주고 fit, 선언하고 전달해주고 fit 이 과정 반복이 사이킷런
```

```python
dt_clf.feature_importances_
# 속성값
# decisiontree가 뭔지는 모르겠지만 제일 중요한건 맨 뒤 (4번째) 거겠네
# -> 꽃잎의 폭, 이게 가장 중요하다.
# 꽃잎의 폭을 가지고 학습을 가장 많이 시킨 것
# (피쳐중에 가장 중요한거 뭐 썼냐)
```

내가 도메인이 없다면, decisionTree 돌려놓고, feature_importances_를 해서 연관성이 가장 높은 상위 3개를 뽑아낼 수 있을 것

학습 알고리즘을 feature selection 단계라고 하는데, 해당 과정을 통해 이용할 수도 있을 것

```python
import matplotlib.pyplot as plt

plt.plot(dt_clf.feature_importances_)
```
